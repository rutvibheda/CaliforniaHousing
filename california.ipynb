Preparing the dataset

```python
dataset = pd.DataFrame(california.data, columns = california.feature_names)
```

```python
dataset.head()
```

```python
dataset['MedianPrice'] = california.target
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
#check for missing values
dataset.isnull().sum()
```

Exploratory Data Analysis

```python
X = dataset.iloc[: , :-1]
y = dataset.iloc[: , -1]
```

```python
X.head()
```

```python
y
```

```python
##Train Test Split
from sklearn.model_selection import train_test_split
X_train , X_test , y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42 )
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_train
```

Model Training

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression.fit(X_train, y_train)
```

```python
#print coefficients and the intercept
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
#on which parameters the model has been trained
regression.get_params()
```

```python
#Prediction with test data
reg_pred = regression.predict(X_test)
```

```python
reg_pred
```

```python
#plot a scatter plot for the prediction
plt.scatter(reg_pred , y_test)
```

```python
#residual means error
residuals = y_test - reg_pred
```

```python
residuals
```

```python
#plotting the residuals

sns.displot(residuals , kind = "kde")
```

```python
#scatter plot wrt prediction and residuals
plt.scatter(reg_pred, residuals)
```

```python
#the above points are evenly distributed - uniformly distributed
#giving an indication that model performs well
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(y_test, reg_pred))
print(mean_squared_error(y_test , reg_pred))
print(np.sqrt(mean_squared_error(y_test, reg_pred)))
```

R square and adjusted R square

```python
from sklearn.metrics import r2_score
score = r2_score(y_test, reg_pred)
print(score)
```

```python
#adjusted R-squared
1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
